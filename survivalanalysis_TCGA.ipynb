#### extract clinical data of TCGA dataset from RTCGA package (survival data)

```r
library(RTCGA)
rownames(infoTCGA())
library(RTCGA.clinical)
library(RTCGA.mRNA)
```

grouping samples based on expression level of a specific herv. mark cancer type in "clin" 

```r
#expr[,which(colnames(expr)=="herv_2965")]
singleherv<-expr[,c(1,which(colnames(expr)=="herv_2965"))]
singleherv[1:4,]
clin <- survivalTCGA(COAD.clinical,extract.cols="admin.disease_code")
#Show the first few lines

dim(clin)
clin[1:4,]
colnames(clin)[2] <- "Sample_ID"
clinsample<-merge(clin,singleherv,by="Sample_ID",all=FALSE)
colnames(clinsample)[5] <- "expr"
dim(clinsample)
clinsample[1:4,]
```

grouping samples based on total herv expression level

```r
#expr[,which(colnames(expr)=="herv_2965")]

therv<-apply(expr[,-1],1,sum)
nexp<-cbind(expr[,1],as.data.frame(therv))
colnames(nexp)[1] <- "Sample_ID"
nexp[1:4,]
clin <- survivalTCGA(BRCA.clinical,extract.cols="admin.disease_code")
#Show the first few lines

dim(clin)
clin[1:4,]
colnames(clin)[2] <- "Sample_ID"
clinsample<-merge(clin,nexp,by="Sample_ID",all=FALSE)
colnames(clinsample)[5] <- "expr"
dim(clinsample)
clinsample[1:4,]
```

```r
library(survival)
library(ggplot2)
library(ggpubr)
library(magrittr)
library(survminer)
```

```r
library(dplyr)
a1<-as.numeric(mean(clinsample$expr))
g1<-clinsample[which(clinsample$expr<a1),]
g1$group=1
g1[1:4,]
g3<-clinsample[which(clinsample$expr>=a1),]
g3$group=3
g3[1:4,]
groupedclin<-as.data.frame(rbind(g1,g3))

# How many samples of each type?
table(groupedclin$group)
# Tabulate by outcome
xtabs(~group+patient.vital_status, data=groupedclin) %>% addmargins()

coxph(Surv(times, patient.vital_status)~group, data=groupedclin)
sfit <- survfit(Surv(times, patient.vital_status)~group, data=groupedclin)
summary(sfit, times=seq(0,365*5,365))
ggsurvplot(sfit, conf.int=TRUE, pval=TRUE)
```
