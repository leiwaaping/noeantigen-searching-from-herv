### calculate peptide region statistic based on 34 paired COAD samples
calculating average depth for every peptide in tumor samples and normal samples "ave_exp"
as well as site wide ave_depth  
also save peptide average depth for ecah sample in "exp" and "expn"  
plot peptide region expression level

```r
#confidence function
confidence<-function(x){
mean<-mean(x)
sigma<-sd(x)
n<-length(x)
ans<-c(mean-sigma*1.96/sqrt(n),mean+sigma*1.96/sqrt(n))
ans
}

for (i in 1:(dim(con)[1])){
    target_herv=as.character(con[i,][[1]])
    start_pos =as.character(con[i,][2])
    end_pos=as.character(con[i,][3])
    unit<-depth[grepl(target_herv, depth$file_ID),]
    
    #extract peptide region
    unit <- unit[unit[,2]>=as.numeric(start_pos), ]
    unit <- unit[unit[,2]<=as.numeric(end_pos), ]
    unit[is.na(unit)] <- 0
    unit<-unit[order(unit[,2]),]
    rownames(unit)<-unit[,2]
    unit <- unit[,-1]
    unit <- unit[,-1]
    
    #see if both o exp in target region
    kk<-as.list(dim(unit))
    kk[1]
    if(kk[1]==0){
        next
    }
    
    #set normal and tumor groups,calculate average
    unit_normal=unit[,seq(2,dim(unit)[2],by=2)]
    unit_tumor=unit[,seq(1,dim(unit)[2],by=2)]
    
    #calculate average expersion in each pos
    normal<- rowMeans(unit_normal)
    tumor<- rowMeans(unit_tumor)
    a<-paste(tumor,sep=",",collapse="; ")
    b<-paste(normal,sep=",",collapse="; ")
    
    if(i==1){
    #tmp=data.frame(herv=0,start_pos=0)
    ave_exp<-data.frame(herv=0,pos=0,ave_exp_in_tumor=0,ave_exp_in_normal=0,tumor=0,normal=0)
    #exp<- cbind(tmp,unit_tumor[1,])
    #expn<- cbind(tmp,unit_normal[1,])
    exp<- unit_tumor[1,]
    expn<- unit_normal[1,]
    i=0
    }
    
    ##calculate average _exp
    #colSums(unit_tumor)
    #t(colSums(unit_tumor)/length(rowSums(unit_tumor)))
    ##output with any character in column, the hold matrix foramt will change to "character", but limma needs "double"
    #exp<-rbind(exp,c(target_herv,start_pos,t(colSums(unit_tumor)/length(rowSums(unit_tumor)))))
    #expn<-rbind(expn,c(target_herv,start_pos,t(colSums(unit_normal)/length(rowSums(unit_normal)))))
    #normalized with peptide-length here
    exp<-rbind(exp,c(t(colSums(unit_tumor)/length(rowSums(unit_tumor)))))
    expn<-rbind(expn,c(t(colSums(unit_normal)/length(rowSums(unit_normal)))))
    ave_exp<-rbind(ave_exp,c(target_herv,start_pos,sum(tumor)/length(tumor),sum(normal)/length(normal),a,b))
    #line <- readLines(con, n = 1)
    
    #rxpression level plotting
    a<-0
    b<-0
    c<-0
    d<-0

    #calculate confidence
    for(j in 1:length(rowMeans(unit_tumor))){
        a<-append(a,confidence(as.numeric(unit_tumor[j,]))[1])
        b<-append(b,confidence(as.numeric(unit_tumor[j,]))[2])
        c<-append(c,confidence(as.numeric(unit_normal[j,]))[1])
        d<-append(d,confidence(as.numeric(unit_normal[j,]))[2])
    }
    
    #save pic
    hervname=strsplit(target_herv,split='_')[[1]][1]
    name=paste(hervname,start_pos,end_pos,sep="_")
    name2=paste("pic/herv",name,".jpeg",sep="")

    jpeg(file=name2)
    plot(tumor,main="peptide expression in tumor(red) and normal(blue) cell",
     ylim=c(0,10),
     col=2,type="l",pch=15,xlab = "herv_pos",ylab ="#read")
    lines(normal,col=4,type="l",pch=15)
    lines(a[-1],col=2,type="l",pch=15,lty=2)
    lines(b[-1],col=2,type="l",pch=15,lty=2)    
    lines(c[-1],col=4,type="l",pch=15,lty=2)
    lines(d[-1],col=4,type="l",pch=15,lty=2) 
    dev.off()
    
}

#unit[1:4,]
#unit_normal[1:4,]
#unit_tumor[1:4,]
exp<-exp[-1,]
expn<-expn[-1,]
ave_exp<-ave_exp[-1,]
dim(ave_exp)
ave_exp[1:4,]
```

### volcano plot

```r
#takes around 30 mins
ttest<-lapply(1:dim(exp)[1],function(x){
    r<-t.test(t(exp[x,]),t(expn[x,]),paired = TRUE)
    mt<-mean(t(exp[x,]))
    mn<-mean(t(expn[x,]))
    if((mt==0)&(mn==0)){
      fc <- 1
    }else if(mt==0){
      fc <- -1024
    }else if(mn==0){
      fc <- 1024
    }else{
      fc <- mt/mn
    }
    re<-c(mt,mn,log(fc,base=2),r$p.value)
    return(re)
  } )
```

```r
library("limma")
library("ggplot2")
library("dplyr")


d<-do.call(rbind, ttest)
d<-as.data.frame(d)
colnames(d)<-c("avg_tumor","avg_normal","logFC","pvalue")
d[is.na(d)]=-10
#plot(d$logFC,-log(d$pvalue,base = 10),pch=16,cex=0.8,xlim=c(-10,10))

logFC_cutoff=2

d<-d%>%mutate(threshold = ifelse(logFC >= logFC_cutoff,"Up", ifelse(logFC<=-logFC_cutoff , "Down", "None")))

this_tile <- paste0('Cutoff for logFC is ',round(logFC_cutoff,1),
                    '\nThe number of up peptides is ',sum(d$threshold == "Up") ,
                    '\nThe number of down peptides is ',sum(d$threshold == "Down")
)

ggplot(data=d, aes(x=logFC, y=-log10(pvalue))) +
  geom_point(alpha=0.4, size=2,aes(colour = threshold)) +
  scale_colour_manual(values = c("Up"= "red", "Down"="blue",  "None"= "black"))+
  theme(legend.position = "none") +
  theme_bw()+ ggtitle( this_tile ) +
  #xlim(c(-12, 12)) + ylim(c(0, 20)) +
  xlab("log2 fold change") + ylab("-log10 p-value")

d<-cbind(ave_exp[,c(1,2)],d)
colnames(d)[1]<-"herv"
colnames(d)[2]<-"start_pos"
d=d[order(-d[,5]),]
dd<-d[1:21,]
dd[order(dd[,1]),]
#d[1:25,]
table(d$threshold)
write.csv(d,file="data/34paired_COAD_peptide_level_statistic.csv",row.names = FALSE,quote = FALSE)
```
